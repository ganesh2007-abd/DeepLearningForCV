# String Tensors

```python
import tensorflow as tf
tensor_string = tf.constant(["hello","World"])
print(tensor_string)
print(tensor_string.numpy())
```

```python
t1 = tf.constant(["My name"," is ","ganesh"])
t2 = tf.strings.join(t1,separator="")
print(t2)
print(t2.numpy())
```

```python
strings = tf.constant(["ganesh","suvarna","ram","Lak"])
lengths = tf.strings.length(strings)
print(lengths)
print(lengths.numpy())
```

```python
strings_lower = tf.strings.lower(["CamelCase","ALLCAPS"])
print(strings_lower)
print(strings_lower.numpy())
```

# Variables

```python
x = tf.constant(9.0)
x_var = tf.Variable(x,name='var1',trainable=False)
print(x_var)
print(x_var.numpy())
```

```python
v = tf.Variable([1,2])
v.assign([6,5])
print(v.numpy())
```

```python
v.assign_add([1,6])
print(v.numpy())
v.assign_sub([1,1])
print(v.numpy())
```

# Controlling Devices

```python
with tf.device('CPU:0'):
  x1 = tf.Variable([1,2])
  x2 = tf.Variable([2,3])
with tf.device('GPU:0'):
  x3 = x1 + x2
print(x3)
print(x3.numpy())
```
